```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as py
```

```python
df = pd.read_csv('zippedData/tmdb.movies.csv.gz')
```

```python
df.head(6)
```

```python
df = df.sort_values(by=['popularity'], ascending=False).head(5)

df
```

```python
ax = df.plot.bar(x='original_title', y='popularity', figsize=(20,10), colormap='autumn', fontsize=11, rot=0);
ax.set_ylabel('Popularity', fontsize=15)
ax.set_xlabel('Movie Titles', fontsize=15)
ax.set_title('Movie Popularity Rankings', fontsize=20)
ax.patch.set_facecolor('none')
ax.patch.set_alpha(0)
```
